## Lemmatization

Have a look at these words: "Played", "playing", "plays", "play".

These words are not entirely unique, as they all basically refer to the root word: "play". Very often, while trying to interpret the meaning of the text using NLP, you'll be concerned about the root meaning and not the tense.

For the algorithms that work based on the number of occurrences of the words, having multiple forms of the same word will reduce the number of counts for the root word. which is 'play' in this case.

**Lemmatization is the method of converting a token to it's root/base form**

Fortunately, spaCy provides a very easy and robust solutions for this and the root form of every token through **Token.lemma_attribut**.

```python
#Lemmatizing the tokens of a doc
for token in my_doc_cleaned:
    print(token.lemma_)
```
