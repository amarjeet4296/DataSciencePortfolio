## Data Preprocessing
Exploring data and observes that, there need to convert ctegorical variables into dummy variables and scale all the values before training the machine learning models.

```python
dataset = pd.get_dummies(df, columns = ['sex', 'cp', 'fbs', 'restecg', 'slope', 'ca', 'thal'])
dataset.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
standardScaler = StandardScaler()
columns_to_scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
dataset[columns_to_scale] = standardScaler.fit_transform(dataset[columns_to_scale])
dataset.head()
```

# Model Fitting

### KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn import metrics
y = dataset['target']
X = dataset.drop(['target'], axis = 1)
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.2,random_state=32)
```

```python
knn= KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train,y_train)
y_pred= knn.predict(X_test)
metrics.accuracy_score(y_test,y_pred)
```

```python
knn_scores = []
for k in range(1, 21):
    knn_classifier =KNeighborsClassifier(n_neighbors = k)
    score = cross_val_score(knn_classifier, X, y, cv = 10)
    knn_scores.append(score.mean())
```

```python
plt.figure(figsize = (10, 10))
plt.plot([k for k in range(1,21)], knn_scores, color = 'red')
for i in range(1,15):
    plt.text(i, knn_scores[i-1], (i, knn_scores[i-1]))
# plt.xticket([i for i in range(1, 15)])
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('K Neighbors Classifier scores for different K values')
plt.show()
```
